```python
#Declararilor starilor din reteaua bayesiana
age = DiscreteDistribution({'>65': 0.53926, '50-65': 0.28302, '15-49':0.16457, '<15':0.01315})
state_age = State(age, name="age")

underlying_medical = DiscreteDistribution({'High': 0.34531, 'Medium':0.32001, 'Low':0.33469})
state_underlying = State(underlying_medical, name="Underlying medical")

sex = DiscreteDistribution({'Male': 0.55924, 'Female': 0.44076})
state_sex = State(sex, name = "sex")

obsesity = DiscreteDistribution({'Yes': 0.12082, 'No': 0.87918})
state_obsesity = State(obsesity, name="obsesity")

other_condition_with = DiscreteDistribution({'False': 0.68327, 'True': 0.31673})
state_other = State(other_condition_with, name="other condition with")

covid19_status = DiscreteDistribution({'Severe': 0.64564, 'Mild': 0.11104, 'Asymptomatic': 0.0, 'None': 0.24331})
state_status = State(covid19_status, name="COVID19 status")

recent_contact_with = DiscreteDistribution({'Yes': 0.75691, 'No': 0.24309})
state_recent = State(recent_contact_with, name="Recent contact with")

multiple_intractions_with = DiscreteDistribution({'False': 0.0, 'True': 1.0})
state_multiple = State(multiple_intractions_with, name="Multiple intractions with")

recent_contact_with2 = DiscreteDistribution({'False': 0.42919, 'True': 0.57081})
state_recent2 = State(recent_contact_with2, name="Recent contact with2")

covid19_or_condition_with_similar = DiscreteDistribution({'Covid19 severe': 0.6456, 'Covid19 mild': 0.111, 'Covid19 asymptomatic': 0.0, 'Non Covid similar symptoms': 0.2433, 'None': 0.0})
state_condition = State(covid19_or_condition_with_similar, name = "COVID19 or condition with similar")

fever = DiscreteDistribution({'no': 0.00, 'yes': 0.99329})
state_fever = State(fever, name="Fever")

length_of_time_of = DiscreteDistribution({'>5 days': 0.80, '<=5 days': 0.20})
state_length = State(length_of_time_of, name="Length of time of")

symptoms_status = DiscreteDistribution({'Deteriorating': 0.0, 'Stable':0.9997, 'Improving': 0.0})
state_symptoms = State(symptoms_status, name="symptoms_status")

covid_alert = DiscreteDistribution({'False': 0.24331, 'True': 0.75669})
state_alert = State(covid_alert, name = "COVID ALERT")

hospitalization_alert = DiscreteDistribution({'False': 0.39464, 'True': 0.60536})
state_hosp = State(hospitalization_alert, name="Hospitalization ALERT")

body_temp = DiscreteDistribution({'=38': 0.0, '>38': 1.0})
state_body = State(body_temp, name="Body Temp")

cough = DiscreteDistribution({'no': 0.0, 'yes': 1.0})
state_cough = State(cough, name="Cough")

fatigue = DiscreteDistribution({'no': 0.51293, 'yes': 0.48707})
state_fatigue = State(fatigue, name="Fatigue")

sputum = DiscreteDistribution({'no':0.71398, 'yes':0.28602})
state_sputum = State(sputum, name="Sputum")

dyspnea = DiscreteDistribution({'no': 0.0, 'yes': 1.0})
state_dyspnea = State(dyspnea, name="Dyspnea")

chills = DiscreteDistribution({'no': 0.46607, 'yes':0.53393})
state_chills = State(chills, name = "Chills")

myalgia = DiscreteDistribution({'no': 0.0, 'yes':1.0})
state_myalgia = State(myalgia, name="Myalgia")

o2_saturation = DiscreteDistribution({'<90': 1.0, '>=90': 0.0})
state_saturation = State(o2_saturation, name= "O2 Saturation")
```

```python
#Analiza subiectului COVID19 cu ajutorul algoritmului Naive Bayes

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
#Incarcarea dataset-ului 
dataset_covid = pd.read_csv("covid19.csv", header = 0)
```

```python
#Afisarea primelor 5 inregistrari din dataset
dataset_covid.head()
```

```python
#Stergerea coloanei referitoare la denumirea tarilor 
dataset_covid = dataset_covid.drop('Country', axis=1) 
```

```python
#Afisarea primelor 5 inregistrari dupa stergerea coloanei "Country"
dataset_covid.head()
```

```python
#Afisarea numarului de linii si de coloane ale dataset-ului
dataset_covid.shape
```

```python
#Generearea statisticilor pe baza dataset-ului
dataset_covid.describe()
```

```python
dataset_covid.isnull().sum()
dataset_covid.isna().sum()
```

```python
#Impartirea dataset-ului in feature si target
feature = dataset_covid.values[:, 1:10]
target = dataset_covid.values[:,0]
```

```python
#Afisarea feature-ului
feature
```

```python
#Afisarea target-ului
target
```

```python
#Impartirea dataset-ului in set de antrenare si de test
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.40, random_state = 100)
```

```python
#Scalarea setului de antrenare si a celui de test si afisarea informatiilor despre acestea
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Antrenarea setului de date si realizarea predictiei
from sklearn.naive_bayes import GaussianNB

clf=GaussianNB()
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
```

```python
#Afisarea mediei si a variatiei fiecarui feature 
print("Means:", clf.theta_)
print("Standard deviations:", clf.sigma_)
```

```python
#Afisarea raportului de clasificare 
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred,zero_division=1))
```

```python
#Calcularea metricii "accuracy"
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy for NB:",accuracy)
```

```python
#Calcularea metricii "recall"
from sklearn.metrics import recall_score
recall=recall_score(y_test,y_pred, average='macro',zero_division=0)
print("Recall for NB:", recall)
```

```python
#Calcularea metricii "precision"
from sklearn.metrics import precision_score
precision=precision_score(y_test,y_pred, average='macro',zero_division=0)
print("Precision for NB", precision)
```

```python
#Calcularea metricii "f-measure"
from sklearn.metrics import f1_score
f_measure=f1_score(y_test,y_pred, average="macro", zero_division=0)
print("F-measure for NB:",f_measure)
```
